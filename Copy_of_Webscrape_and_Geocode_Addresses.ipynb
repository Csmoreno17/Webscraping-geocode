# A Giant For Loop (for Alabama)
Now let's scrape EVERY TOWN for Alabama...

This is the hard-coded link for one State (Alabama).

```python
# this is alabama
tmp_url = 'http://cvs.com/store-locator/cvs-pharmacy-locations//New-Mexico'
```

Now we can get every link off of this webpage (open it on the side if you want.)

```python
# https://www.geeksforgeeks.org/extract-all-the-urls-from-the-webpage-using-python/

import requests
from bs4 import BeautifulSoup

# url is given above,too
url = tmp_url
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

# let's not just print them, but store them too
urls = []
for link in soup.find_all('a'):
  print(link.get('href'))
  x = link.get('href') 
  urls.append(x)
```

Clearly there are a bunch of stores in Alabama! Now let's grab all of the locations in each of these towns (and drop the tracking/session cookie, too.)

```python
import fnmatch
l = res
pattern = '/store-locator/cvs-pharmacy-locations/*' # note the wildcard!
matching = fnmatch.filter(l, pattern)
print(matching)
print(np.shape(matching)) #101 stores in New Mexico
```

Drop the cookie!

```python
# got it! 
matching = [i.split(';', 1)[0] for i in matching]
print(matching)
print(np.shape(matching)) #101 stores in New Mexico
```

Make the path.

```python
# https://stackoverflow.com/questions/2050637/appending-the-same-string-to-a-list-of-strings-in-python
matching = ['http://cvs.com' + s for s in matching]
matching
```

Now, for each of these 101... go digging!

```python
# rename 'matching' since we use this name later
websiteList = matching
```

```python
results = []
for a in websiteList:
  print(a) # this is just the website
  a_website = urllib.request.urlopen(a)
  a_soup = bs4.BeautifulSoup(a_website)
  website_text = a_soup.findAll(text = True) # grab the website text from that town

  # now pull the first wildcard
  l = website_text
  pattern = '\n\t\t\t\t\t\t\t*' # note the wildcard!
  matching = fnmatch.filter(l, pattern)
  print('raw:', np.shape(matching)) # this should be a big number

  # now pull the second wildcard
  # we are overwriting the name 'matching' alot, but it's OK
  l = matching
  pattern = '*\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\t\t\n\t\t\t\t\t\t\t\n\t\t\t\t\t\t' # note the wildcard!
  matching = fnmatch.filter(l, pattern)
  print(matching)
  print('clean:', np.shape(matching)) # 2 entries - it worked!

  # now append
  results.append(matching)

# there should be 101 stores? more?
print(np.shape(results))
```

```python
locator = Nominatim(user_agent='YourNetID_here')

# here's a sample location for CVS in New Mexico!
location = locator.geocode('8510 Montgomery Blvd Ne Albuquerque, NM 87111')
print("Latitude = {}, Longitude = {}".format(location.latitude, location.longitude))
```

```python
# from tmp back to df...
df = tmp
```

The address must be a single string (column). This is where you will need to update your script for new data (potentially).

```python
from geopy.extra.rate_limiter import RateLimiter

# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column (THIS IS THE ONE YOU MAY NEED TO UPDATE - ADDRESS MUST BE SINGLE STRING)
df['location'] = df['Address'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)
```

It should take one second per record to geocode... try to update the min_delay_seconds if it doesn't work.
